Сделали файл инпута

```python
with open('o_henry.conllu') as f, open('input.conllu','w') as inp:
    for line in f:
        if line == '' or line == '\n':
            continue
        elif line.startswith('#'):
            inp.write(line)
        else:
            fields = [line.split('\t')[0],line.split('\t')[1]]
            for i in range(8):
                fields.append('-')
            print('\t'.join([field for field in fields]), file = inp)
```

```python
file = open('model.tsv')
word_info = {} #{word: {tag:tag_count}}
for line in file:
    
    if line.startswith('#'):
        continue
    
    word = line.split('\t')[3].strip()
    tag = line.split('\t')[2]
    count_tag = int(line.split('\t')[1])

    #if word == '-':
        #continue

    if not word in word_info:
        word_info[word]= {}
        word_info[word][tag] = 0
    elif tag not in word_info[word]:
        word_info[word][tag] = 0
    word_info[word][tag] += count_tag
```

```python
word_info
```

```python
word_tag = {}
for word, tags in word_info.items():
    word_tag[word] = max(tags, key=tags.get)
```

```python
word_tag
```

```python
def tagger(line, dictionary):
    line = line.split('\t')
    word = line[1]
    tag = dictionary[word]
    line[3] = tag
    new_line = ('\t').join([field for field in line])
    return new_line
```

```python
text = open('input.conllu')
out = open('output.conllu','w')
for line in text:
    line = line.strip()
    if line.startswith('#'):
        print(line, file = out)
    else:
        print(tagger(line, word_tag), file = out)
```

```python
with open('o_henry.conllu') as inp, open('output.conllu') as out:
    mistakes = 0
    word_tag_inp = [(line.split('\t')[1], line.split('\t')[3]) for line in inp if not line.startswith('#') and line != '\n']
    tags_out = [line.split('\t')[3] for line in out if not line.startswith('#') and line != '\n']
    total = len(tags_out)
    for i in range(total):
        if word_tag_inp[i][1] != tags_out[i]:
            mistakes +=1
            print(word_tag_inp[i][0], tags_inp[i], tags_out[i], sep = '\t')
    print((total-mistakes)*100/total)
```
