```python
acd = "https://yle.fi/uutiset/"
index = -1
url_name = acd.split('/')[index]
if (url_name ==''):
    url_name =  acd.split('/')[index - 1]
url_name
```

```python
def read_article(url):
    req = requests.get(url).text
    soup = BeautifulSoup(req,"lxml")
    text_items = soup.findAll('p')
    save_to_file = []
    #add_name = True
    for item in text_items:
        #print(add_name)
        save_to_file.append(item.text)
        
    index = -1
    name = url.split('/')[index] 
    if (name ==''):
        name =  acd.split('/')[index - 1] + '.txt'
    else:
        name += '.txt'
    onlyfiles = [f for f in listdir("C://input//fin") if isfile(join("C://input//fin", f))]
    if (name in onlyfiles):
        name = 'new_' + name
    file_name = "C://input//fin//" + name  
    with io.open(file_name, "w+") as newsfile:
        for line in save_to_file:
            newsfile.write(line + '\n')
    print('saved with name = ', name)
```
