```python
def read_article(url):
    req = requests.get(url).text
    soup = BeautifulSoup(req,"lxml")
    text_items = soup.findAll('p')
    save_to_file = []
    #add_name = True
    for item in text_items:
        #print(add_name)
        save_to_file.append(item.text)
    
    print(url.split('/'))
  
    index = -1
    
    name = url.split('/')[index] 
    if (name ==''):
        name =  url.split('/')[index - 1] + '.txt'
    else:
        name += '.txt'
        
    onlyfiles = [f for f in listdir("C://input//fin_news") if isfile(join("C://input//fin_news", f))]
    if (name in onlyfiles):
        print("DUPLICATED ARTICLE FOUND")
        remove("C://input//fin_news//" + name) 
    file_name = "C://input//fin_tosi_easy_news//" + name  
    with io.open(file_name, "w+",encoding ="utf-8") as newsfile:
        for line in save_to_file:
            newsfile.write(line + '\n')
    print('saved with name = ', name)
```
